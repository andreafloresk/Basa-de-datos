```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

adultpath = '/content/drive/My Drive/a4/adult.csv'
antropath = '/content/drive/My Drive/a4/antropometria-dataset-2.csv'

df1=pd.read_csv(adultpath)
df2=pd.read_csv(antropath)

df1.head(5)
```

```python
favs=["entidad","peso","sexo","desc_ent"]
ndf2=df2[favs]
ndf2
```

```python
pd.pivot_table(ndf2,index="desc_ent",aggfunc="mean", values="peso")
```

```python
pd.pivot_table(ndf2,values="peso",aggfunc="mean", index="desc_ent",columns="sexo")
```

```python
pivot_table = pd.pivot_table(df1, index='marital-status', aggfunc='size')
print(pivot_table)
```

```python
pivot_table1 = df1.pivot_table(index='marital-status', columns='education', values='hours-per-week', aggfunc='mean')
print(pivot_table1)
```

```python
favs1=["education","marital-status","hours-per-week","age","sex"]
ndf1=df1[favs1]
ndf1
```

```python
pd.pivot_table(ndf1, index=['marital-status', 'education'], columns='sex', values=['hours-per-week', 'age'], aggfunc="median")

```
