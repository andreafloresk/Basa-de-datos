## A5 Visualización con histogramas y boxplots

```python
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

path = "/content/drive/My Drive/a4/antropometria-dataset-1.csv"

df=pd.read_csv(path)
df.head(5)
```

```python
df_clean = df.dropna(subset=['cintura'])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot(df_clean['peso'])
plt.title('Gráfico de caja del Peso (peso)')
```

```python
plt.subplot(1, 2, 2)
plt.boxplot(df_clean['cintura'])
plt.title('Gráfico de caja de la Cintura (cintura)')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df_clean['peso'])
plt.title('Gráfico de línea del Peso (peso)')
plt.xlabel('Índice')
plt.ylabel('Peso (kg)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
plt.hist(df_clean['peso'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma del Peso (peso)')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')
```

```python
plt.subplot(1, 2, 2)
plt.hist(df_clean['cintura'], bins=30, color='salmon', edgecolor='black')
plt.title('Histograma de la Cintura (cintura)')
plt.xlabel('Cintura (cm)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
df_sex1 = df_clean[df_clean['sexo'] == 1]
df_sex2 = df_clean[df_clean['sexo'] == 2]
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df_sex1['peso'])
plt.title('Gráfico de caja del Peso (peso) para Sexo 1')
```

```python
plt.subplot(1, 2, 2)
plt.boxplot(df_sex2['peso'])
plt.title('Gráfico de caja del Peso (peso) para Sexo 2')
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df_sex1['peso'], bins=30, color='blue', edgecolor='black')
plt.title('Histograma del Peso (peso) para Sexo 1')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')
```

```python
plt.subplot(1, 2, 2)
plt.hist(df_sex2['peso'], bins=30, color='pink', edgecolor='black')
plt.title('Histograma del Peso (peso) para Sexo 2')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['peso'], df_clean['cintura'], alpha=0.5)

plt.title('Diagrama de dispersión entre Peso y Cintura')
plt.xlabel('Peso (kg)')
plt.ylabel('Cintura (cm)')
plt.grid(True)
plt.show()
```
